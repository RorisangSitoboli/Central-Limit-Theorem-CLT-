<b>Central Limit Theorem (CLT)</b>
- Given a dataset with an unknown distribution, the sample's means will approximate the normal distribution.
- The CLT places no constraints on the distribution of the function used for the generation of the N samples, from which the random variable is formed. 
- It is not important what the distribution associated with the random variable is, but when the average is characterized by a finite variance and is obtained for a very <b>large number of samples</b>, it can be described through a Gaussian
distribution.

```python
# Import some libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
a = 1
b = 100    # a & b define the range we'll use to generate the pseudo-random numbers.
N = 10000  # Generate 10 000 random numbers from a uniform distribution
Pop = list(np.random.uniform(a, b, N))
```

```python
# Plot a histogram showing the density of the population/distribution
plt.figure(figsize = (10, 5))
plt.hist(Pop, density = True, histtype = 'stepfilled', alpha = 0.5)
plt.xlabel('Sample', fontsize = 15)
plt.ylabel('Density', fontsize = 15)
plt.title('Uniformly distributed population', fontsize = 15)
plt.grid()
plt.show()
```

```python
# Now let's plot the means of each sample.
Samples_Means = [] # Empty list to hold the means.
for i in range(0, 1000): # Run the loop a 1000 times.
    Extracted_Data = random.sample(Pop, k = 100)
    Extracted_Data_Mean = np.mean(Extracted_Data)
    Samples_Means.append(Extracted_Data_Mean)
plt.figure(figsize = (10, 5))
plt.hist(Samples_Means, density = True, histtype = 'stepfilled', alpha = 0.6)
plt.xlabel('Sample mean', fontsize = 15)
plt.ylabel('Density', fontsize = 15)
plt.title('Profile of the means of the samples', fontsize = 15)
plt.grid()
plt.show()
```

<b>Summary</b>
- The distribution is now a typical bell-shaped/S-shaped.
- The CLT has been proved.
